```python
import joblib
import requests
```

```python
model = joblib.load('../Models/fake_news_model.pkl')
vectorizer = joblib.load('../Models/bow_vectorizer.pkl')
print("Model and vectorizer loaded successfully.")
```

```python
def predict_news(text):
    new_vec = vectorizer.transform([text])
    prediction = model.predict(new_vec)[0]
    if prediction ==1:
        return "Real News"
    else:
        return "Fake News"
```

```python
def fact_check(news_text):

    url = "https://factchecktools.googleapis.com/v1alpha1/claims:search"

    params = {
        "query": news_text,
        "languageCode": "en",
        "key": "AIzaSyAE56OYa_6G-Ocuzr_bbauwulEdww6McR0"
    }

    response = requests.get(url, params=params, verify=False)

    print(response.status_code)
    print(response.text)

    if response.status_code != 200:
        return None, None

    data = response.json()

    if "claims" in data:

        claim = data["claims"][0]

        rating = claim["claimReview"][0]["textualRating"]
        source = claim["claimReview"][0]["publisher"]["name"]

        return rating, source

    return None, None
```

```python
def detect_fake(news_text):
    ml_prediction = predict_news(news_text)
    rating,source = fact_check(news_text)
    print(f"ML Prediction: {ml_prediction}")
    if rating:
        print(f"Fact Check Rating: {rating} (Source: {source})")
        print("Final Result: Verified using Fact-check source")
    else:
        print("No fact check result found")
        print("Final result:",ml_prediction)
```

```python
news = "Donald Trump suggested injecting disinfectant to treat coronavirus"
detect_fake(news)
```
