```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv(r'C:\Users\hp\OneDrive\Documents\Desktop\fake news detection\Data\cleaned_data.csv')
df.head()
```

```python
df['label'].value_counts()
```

```python
sns.countplot(x='label', data=df)
plt.title('Distribution of Real and Fake News')
plt.show()
```

```python
df['text_length'] = df['clean_content'].astype(str).apply(len)
sns.histplot(data=df, x='text_length', hue='label', bins=50)
plt.title('Distribution of Text Length by Label')
plt.xlabel('Text Length')
plt.ylabel('Frequency')
plt.show()
```

# Word Cloud for all News

```python
all_words = " ".join(df['clean_content'].astype(str))

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(all_words)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Word Cloud for All News")
plt.show()
```

# For Fake News

```python
fake_news = df[df['label'] == 0]

fake_words = " ".join(fake_news['clean_content'].astype(str))

wordcloud_fake = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(fake_words)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud_fake)
plt.axis("off")
plt.title("Word Cloud for Fake News")
plt.show()
```
